# Applying RandomSearchCV 

### Hyperparameter Tuning

```python
RF = RandomForestClassifier() #call the RF model
```

```python
### parameters
parameters = {'max_depth':np.arange(1,10),
               'min_samples_split': np.arange(2,5), 
               'random_state':[3], 
               'n_estimators': np.arange(20,30)}
```

```python
parameters
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
### Define search
search = RandomizedSearchCV(RF, parameters, scoring='accuracy')
```

```python
### Execute the search
result = search.fit(x_train, y_train)
```

```python
result.best_score_
```

```python
best_params = result.best_params_
best_params
```

```python
rf1 = RandomForestClassifier(**best_params)
rf1.fit(x_train, y_train)
```

```python
y_pred = rf1.predict(x_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

### Conclusions:

 #1. Random Forest Classifier - without removing outliers after VIF - 85%
 #2. Random Forest Classifier - by removing outliers after VIF - 100%. However, it gives only 1 value in the confusion matrix.
 
 It is not advisable to remove outliers in such kind of data, where the columns have categorical information.
